6. If/else statements - Exercise 

```python
import matplotlib.pyplot as plt  # import the matplotlib package

x_min, x_max = -5, 8

def clip_list(x, x_min, x_max):
    """Clip the values in a list so that they fall within the given max and min values."""
    x_clip = []  # Create an empty list to store the clipped values
    for value in x:
        if value < x_min:
            x_clip.append(x_min)  # If value is smaller than x_min, set it to x_min
        elif value > x_max:
            x_clip.append(x_max)  # If value is larger than x_max, set it to x_max
        else:
            x_clip.append(value)  # If value is within the range, keep it as is
    return x_clip


def plot_lines(x, y_unclipped, y_clipped, label1, label2):
    """Plot the unclipped and clipped lines."""
    plt.plot(x, y_unclipped, label=label1, linestyle='-', color='blue')  # Unclipped line
    plt.plot(x, y_clipped, label=label2, linestyle='--', color='red')   # Clipped line
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.title('Unclipped vs Clipped Values')
    plt.grid(True)
    plt.show()

# Example usage
x = range(-8, 10)  # x-values that we want to calculate and plot
a_line, b_line = 0.5, 2  # Example line parameters for y = ax + b

y_unclipped = evaluate_line_list(a_line, b_line, x)  # Get the unclipped values
y_clipped = clip_list(y_unclipped, -5, 5)  # Clip the values within the range [-5, 5]

# Plot the lines
plot_lines(x, y_unclipped, y_clipped, label1='Unclipped', label2='Clipped')
```

7. Basics of Numpy

```python
import numpy as np
import matplotlib.pyplot as plt

def evaluate_line_numpy(a, b, x):
    """Evaluate a line y = a*x + b using NumPy."""
    x = np.array(x)  # Convert x to a numpy array in case a list was passed in
    y = a * x + b  # Evaluate line
    return y

def evaluate_parabola_numpy(a, b, c, x):
    """Evaluate a parabola y = ax^2 + bx + c using NumPy."""
    x = np.array(x)  # Convert x to a numpy array if it isn't already
    y = a * x**2 + b * x + c  # Evaluate parabola
    return y

def plot_lines(x, y_unclipped, y_clipped, label1, label2):
    """Plot the unclipped and clipped lines."""
    plt.plot(x, y_unclipped, label=label1, linestyle='-', color='blue')  # Unclipped line
    plt.plot(x, y_clipped, label=label2, linestyle='--', color='red')   # Clipped line
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.title('Unclipped vs Clipped Values')
    plt.grid(True)
    plt.show()

# Inputs to line function
a_line, b_line = 2, 1  # Example line parameters for y = ax + b

# Inputs to parabola function
a_parabola, b_parabola, c_parabola = 1, -2, 1  # Example parabola parameters for y = ax^2 + bx + c

# x-values that we want to calculate and plot
x = np.linspace(-10, 10, 100)

# Evaluate the line and parabola using NumPy
y_line_unclipped = evaluate_line_numpy(a_line, b_line, x)  # Unclipped line values
y_parabola_unclipped = evaluate_parabola_numpy(a_parabola, b_parabola, c_parabola, x)  # Unclipped parabola values

# Clip the values to the range [-5, 5]
y_line_clipped = np.clip(y_line_unclipped, -5, 5)  # Clipped line values
y_parabola_clipped = np.clip(y_parabola_unclipped, -5, 5)  # Clipped parabola values

# Plot the lines and the parabola
plot_lines(x, y_line_unclipped, y_line_clipped, label1='Line Unclipped', label2='Line Clipped')
plot_lines(x, y_parabola_unclipped, y_parabola_clipped, label1='Parabola Unclipped', label2='Parabola Clipped')
```
